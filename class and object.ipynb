```python
class Restaurant:
    def __init__(self, item, orders, bill, GST):
        self.item = item
        self.orders = orders
        self.bill = bill
        self.GST = GST

    def payment(self):
        print("Order item:", self.item)
        print("Total bill for item:", self.bill)

        if self.orders > 1:
            print("GST:", self.GST * self.orders)
        else:
            print("GST:", self.GST)

        totalamount = (self.orders * self.bill) + self.GST
        print("Total amount:", totalamount)


class OnlineDelivery(Restaurant):

    def __init__(self, item, orders, bill, GST,
                 onlineorders, deliverycharges, distance):

        super().__init__(item, orders, bill, GST)

        self.onlineorders = onlineorders
        self.deliverycharges = deliverycharges
        self.distance = distance

    def total_bill(self):

        if self.distance > 5:
            self.deliverycharges = self.distance * 10

        else:
            self.deliverycharges = self.distance * self.deliverycharges

        total_amount = (
            self.onlineorders * self.bill
            + self.GST
            + self.deliverycharges
        )
        print("Item name:",self.item)
        print("no.of orders:",self.onlineorders)
        print("Total amount:",self.bill)
        print("GST:", self.GST)
        print("Delivery charges:",self.deliverycharges)

        print("Total payment:", total_amount)


Bill = OnlineDelivery(
    "biriyani",
    2,
    200,
    20,
    2,
    5,
    5
)

Bill.total_bill()
Bill.payment()
        
        
```

```python
class StockTradingAccount:

    def __init__(self, account_name, shares):
        self.account_name = account_name
        self.__shares = shares

    def trading(self):

        while True:

            choice = int(input("""
1.Buy Shares
2.Sell Shares
3.Exit

Enter Choice: """))

            if choice == 1:

                qty = int(input("Enter quantity to buy: "))
                self.__shares += qty
                print("Shares Purchased Successfully")

            elif choice == 2:

                qty = int(input("Enter quantity to sell: "))

                if qty <= self.__shares:
                    self.__shares -= qty
                    print("Shares Sold Successfully")

                else:
                    print("Cannot sell more than owned shares")

            elif choice == 3:
                print("Thank You")
                break

            else:
                print("Invalid Option")

    def get_details(self):
        print("Account Holder:", self.account_name)
        print("Available Shares:", self.__shares)


acc = StockTradingAccount("Prasad", 100)

acc.trading()
acc.get_details()
```

```python
class LoanSystem:

    def __init__(self, name, loan_amount, tenure, interest):
        self.name = name
        self.loan_amount = loan_amount
        self.tenure = tenure
        self.interest = interest
        self.__emi = 0

        self.calculate_emi()     # automatic calculation

    def calculate_emi(self):

        total = self.loan_amount + (
            self.loan_amount * self.interest * self.tenure
        ) / 100

        self.__emi = total / (self.tenure * 12)

    def change_tenure(self, new_tenure):

        self.tenure = new_tenure
        self.calculate_emi()     # automatic recalculation

    def get_details(self):

        print("Customer:", self.name)
        print("Loan Amount:", self.loan_amount)
        print("Tenure:", self.tenure, "Years")
        print("EMI:", round(self.__emi,2))


loan = LoanSystem("Prasad",100000,2,10)

loan.get_details()

print("\nChanging Tenure...\n")

loan.change_tenure(4)

loan.get_details()
        
        
```

```python
class BankLocker:

    def __init__(self, password):

        self.__password = password
        self.__attempts = 0
        self.__locked = False

    def access_locker(self):

        if self.__locked:
            print("Locker Permanently Locked")
            return

        entered = input("Enter Password: ")

        if entered == self.__password:

            print("Access Granted")
            self.__attempts = 0

        else:

            self.__attempts += 1
            print("Wrong Password")

            if self.__attempts >= 3:
                self.__locked = True
                print("Locker Permanently Locked")


locker = BankLocker("1234")

locker.access_locker()
locker.access_locker()
locker.access_locker()
locker.access_locker()
```
